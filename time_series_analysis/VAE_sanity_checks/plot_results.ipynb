# VAE Sanity Checks — Results

Notebook for the visual inspection of the VAE training results.
To be used after running `VAE_sanity_checks_TrainOnSyntheticData.ipynb` (or `.py`).

### Input
    ./history_files/{config_name}_{seed}.csv     training history per seed
    ./weights/{config_name}_{seed}_init.npz      weights before training
    ./weights/{config_name}_{seed}_trained.npz   weights after training

### Workflow
1. Plot the training and validation loss histories for all seeds.
2. Visually identify which seeds converged and which did not,
   and encode the result in the `successes` array.
3. Compare the weight distributions before and after training
   via violin plots and histograms, separately for converged and failed runs.
4. Inspect the element-wise weight differences (trained − init)
   to characterise how much each run moved during training.

### Dependencies
`weights_tools.py`

### Notes
- `dataset_name` and `seeds` must match the values used during training.
- The `successes` array is filled manually after visual inspection of the loss curves.

```python
import os
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from weights_tools import *

seeds = np.array([32467, 75410, 87278, 34987, 38956, 26012, 67418, 47842, 66012, 87346])
dataset_name = "VAE_4dim_7000evts_10good-per-batch"
```

```python
# Plot histories 

for i in seeds:
    df = pd.read_csv(f'./history_files/{dataset_name}_{i}.csv')
    plt.plot( df['epoch'], df['loss'], label=f'{i}' )

plt.ylim(0,20)
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend()
plt.show()
```

```python
# Violin plots of the weights for one layer, separately before and after the training, for all the seeds
plot_violins(dataset_name, seeds, phase="trained",
             submodel="encoder", layer="conv1", weight_type="kernel")
```

```python
# Histograms of the weights for one layer, separately before and after the training, for all the seeds
plot_histograms(dataset_name, seeds, phase="trained", 
                submodel="encoder", layer="conv1", weight_type="kernel")
```

```python
# Compare weights at different training phases for a single seed

plot_violin_comparison(dataset_name, seed=seeds[0], 
                       submodel="encoder", layer="conv1", weight_type="kernel")

plot_histograms_comparison(dataset_name, seed=seeds[0], 
                           submodel="encoder", layer="conv1", weight_type="kernel")
```

#### After checking the histories, one can evaluate which models successfully converged, and which failed the training.
#### Then one can built a simple boolean array to be used in the weights plotting functions.

```python
successes = [0,1,0,0,1,1,1,0,0,1]
```

```python
# Violin plots of the weights for one layer, separately before and after the training, for all the seeds
plot_violins(dataset_name, seeds, phase="init", success=successes,
             submodel="encoder", layer="conv1", weight_type="kernel")
```

```python
# Histograms of the weights for one layer, separately before and after the training, for all the seeds
plot_histograms(dataset_name, seeds, phase="init", success=successes, 
                submodel="encoder", layer="conv1", weight_type="kernel")
```

```python
# Plot histogram of the weights differences at different training phases
plot_histograms_deltas(dataset_name, seeds, success=successes,
                       submodel="encoder", layer="conv1", weight_type="kernel")
```
