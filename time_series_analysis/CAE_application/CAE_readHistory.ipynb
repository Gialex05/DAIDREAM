## Training History - CAE

Reads the training log produced by `CAE_training.py` and visualizes
the loss and learning rate history, with each model/round shown as a
separate line.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Parameters
`epochs_per_training` must match the value set in CAE_training.py.
Check also the name of the CSV files containing the information.

```python
df = pd.read_csv('CAE.csv')

# Identify the start of each new model/round as rows where
# epoch does not increase with respect to the previous row

epochs_per_training = 5  # must match the value in CAE_training.py

split_mask = (df['epoch'].diff() <= 0) | (df['epoch'] % epochs_per_training == 0)
split_indices = sorted(set([0] + list(df.index[split_mask]) + [len(df)]))
```

### Loss history (train and validation)

```python
fig, ax = plt.subplots(figsize=(10, 4))

for k in range(len(split_indices) - 1):
    segment = df.iloc[split_indices[k]:split_indices[k+1]]
    ax.plot(segment['epoch'], segment['loss'],     color='red',  lw=1.5, alpha=0.7)
    ax.plot(segment['epoch'], segment['val_loss'], color='blue', lw=2, linestyle='--', alpha=0.4)

# Add legend with dummy lines
ax.plot([], [], color='red',  lw=1.5, label='train')
ax.plot([], [], color='blue', lw=2, linestyle='--', label='validation')

ax.set_xlabel('epochs')
ax.set_ylabel('loss')
ax.legend()

plt.ylim(0,5)
plt.tight_layout()
plt.show()
```

### Learning rate history

```python
fig, ax = plt.subplots(figsize=(10, 4))

for k in range(len(split_indices) - 1):
    segment = df.iloc[split_indices[k]:split_indices[k+1]]
    ax.plot(segment['epoch'], segment['lr'],     color='orange',  lw=1, alpha=0.8)

# Add legend with dummy lines
ax.plot([], [], color='orange',  lw=1, label='l.r.')

ax.set_xlabel('epochs')
ax.set_ylabel('l.r.')
ax.legend()

#plt.ylim(0,5)
plt.tight_layout()
plt.show()
```
